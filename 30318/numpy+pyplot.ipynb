```python
t=np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)

plt.plot(t,y1,'r--',marker='*')
plt.plot(t,y2,'go-')


plt.grid()
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1,'r--',marker='*')
plt.subplot(1,2,2)
plt.plot(t,y2,'go-')
```

```python
x=np.arange(0,5,1)
x1=x+1
x2=x+2
x3=x+3
x4=x+4
```

```python
plt.subplot(1,2,1)
plt.plot(x,x1,'y--',marker='^',label='chicken')
plt.plot(x,x2,'b-')
plt.ylabel('kor')
plt.xlabel('ea')

plt.subplot(1,2,2)
plt.plot(x,x3,'y--',marker='^',label='chicken')
plt.plot(x,x4,'b-')
plt.ylabel('hell')
plt.xlabel('o')
```

```python
t=df.head(5)
t
```

```python
a=t['이름']
b=t['신해삼']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(a,b)

plt.ylabel('이름')
plt.xlabel('영어')
plt.title('chicken')
plt.legend('value')
```
